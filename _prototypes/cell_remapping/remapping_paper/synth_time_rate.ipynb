```python
def _interpolate_matrix(matrix, new_size=(257,257), cv2_interpolation_method=cv2.INTER_CUBIC):
    '''
        Interpolate a matrix using cv2.INTER_LANCZOS4.
    '''
    return cv2.resize(matrix, dsize=new_size,
                      interpolation=cv2_interpolation_method)
```

```python
def make_single_field(N=16, stdev=10, shift_row=0, shift_col=0, custom_gauss=None):
    """
    N is kernel size
    """
    if custom_gauss is None:
        k1d = signal.gaussian(N, std=stdev).reshape(N, 1)
        kernel = np.outer(k1d, k1d)
    else:
        kernel = custom_gauss
    single_field = np.zeros((N*3, N*3))
    single_field[N+ceil(N/2)+shift_row, N+ceil(N/2)+shift_col] = 1  
    row, col = np.where(single_field == 1)
    rowend = row[0]+ceil(N/2)
    colend = col[0]+ceil(N/2)
    if colend >= len(single_field):
        colend -= len(single_field)
        colinter = N - colend
    else:
        colinter = N
    if rowend >= len(single_field):
        rowend -= len(single_field)
        rowinter = N - rowend
    else:
        rowinter = N

    # if N % 2 == 0:
    #     single_field[row[0]-ceil(N/2):row[0]+ceil(N/2), col[0]-ceil(N/2):col[0]+ceil(N/2)] = kernel[:rowinter, :colinter]
    # else:
    #     single_field[row[0]-N//2:row[0]+ceil(N/2), col[0]-N//2:col[0]+ceil(N/2)] = kernel[:rowinter, :colinter]

    single_field[row[0]-ceil(N/2):row[0]+N//2, col[0]-ceil(N/2):col[0]+N//2] = kernel
    return single_field

N = 17
stdev = 3
single_field = make_single_field(N, stdev)
k1d = signal.gaussian(N, std=stdev).reshape(N, 1)
kernel = np.outer(k1d, k1d)

fig = plt.figure(figsize=(8,5))
ax = plt.subplot(1,2,1)
ax.imshow(_interpolate_matrix(kernel), cmap='jet')
ax = plt.subplot(1,2,2)
ax.imshow((single_field), cmap='jet')
rect = patches.Rectangle((N, N), N-1, N-1, linewidth=1, edgecolor='r', facecolor='none')
ax.add_patch(rect)
fig.tight_layout()
fig.suptitle('Place field (left) and (3*N,3*N) map for easy manipulation of place field')
plt.show()

```

```python
import cv2

def _gkern(kernlen: int, std: int) -> np.ndarray:

    '''
        Returns a 2D Gaussian kernel array.

        Params:
            kernlen, std (int):
                Kernel length and standard deviation

        Returns:
            np.ndarray:
                gkern2d
    '''

    gkern1d = signal.gaussian(kernlen, std=std).reshape(kernlen, 1)
    gkern2d = np.outer(gkern1d, gkern1d)
    return gkern2d
```

```python
from scipy.stats import pearsonr, spearmanr, wasserstein_distance
from _prototypes.cell_remapping.src.wasserstein_distance import _get_ratemap_bucket_midpoints, single_point_wasserstein
import time
import timeit

source_collection_2d = []
target_collection_2d = []
# for j in range(100):
for size in np.arange(2,64*3,2):
    source_collection_2d.append(np.zeros((size,size)))
    target_collection_2d.append(np.zeros((size,size)))


pearson_collection_2d = []
sliced_emd_collection_2d = []
sliced_emd_collection_2d_2 = []
single_point_emd_collection_2d = []

def wass_sliced_fxn(coord_buckets, source_weights, target_weights, n_proj):
    emd_wass = pot_sliced_wasserstein(coord_buckets, coord_buckets, source_weights, target_weights, n_projections=n_proj)
    return emd_wass


for j in range(1):
    for i in range(len(source_collection_2d)):
        source = source_collection_2d[i]
        target = target_collection_2d[i]


        y, x = source.shape
        # if y % 2 == 0:
        #     # even
        #     pass
        # else:
        #     # odd
        #     y = y - 1
        assert y % 2 == 0, 'y must be even'
        height_bucket_midpoints, width_bucket_midpoints =  _get_ratemap_bucket_midpoints(([1],[1]), y, x)

        coord_buckets = np.array(list(itertools.product(np.arange(0,y,1),np.arange(0,x,1))))
        source_weights = np.array(list(map(lambda x: source[x[0],x[1]], coord_buckets)))
        target_weights = np.array(list(map(lambda x: target[x[0],x[1]], coord_buckets)))
        source_weights = source_weights / np.sum(source_weights)
        target_weights = target_weights / np.sum(target_weights)


        prson_fxn = lambda x,y: lambda: pearsonr(x,y)
        sliced_fxn = lambda z,x,y,w: lambda: wass_sliced_fxn(z, x, y, w)
        def spw_fxn(source, width_bucket_midpoints, height_bucket_midpoints):
            spw = np.sum(list(map(lambda x, y: np.linalg.norm(np.array((0, 0)) - np.array((width_bucket_midpoints[x],height_bucket_midpoints[y]))) * source[x,y], np.arange(len(source)), np.arange(len(source)))))
            return spw
        

        elapsed_time = timeit.timeit(prson_fxn(source.flatten(),target.flatten()), number=100)
        pearson_collection_2d.append(np.mean(elapsed_time))

        elapsed_time = timeit.timeit(sliced_fxn(coord_buckets, source_weights, target_weights, 10**2), number=100)
        sliced_emd_collection_2d.append(np.mean(elapsed_time))

        elapsed_time = timeit.timeit(sliced_fxn(coord_buckets, source_weights, target_weights, 10**3), number=100)
        sliced_emd_collection_2d_2.append(np.mean(elapsed_time))

        elapsed_time = timeit.timeit(lambda: spw_fxn(source, width_bucket_midpoints, height_bucket_midpoints), number=100)
        single_point_emd_collection_2d.append(np.mean(elapsed_time))

        
```
